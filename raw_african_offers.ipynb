```python
#scraping data from raw-african website to check for new offers 



#import libraries 


from bs4 import BeautifulSoup
import requests
import pandas as pd
import datetime
```

```python
# connecting with the website



URL = "https://rawafrican.net/collections/offers"


HEADERS = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/116.0"}


page = requests.get(URL, headers=HEADERS)


soup = BeautifulSoup(page.content, "html.parser")
```

```python
# Function to get the product name


def get_name(i):
    
        name = soup.findAll("h3", {"class": "product-title"})     
        name_value = name[i].text  

        return name_value
    
    

    
    
# Function to get the product category


def get_category(i):

        category = soup.findAll("p", {"class": "wbprotype"})
        category_value = category[i].text  
        
        if (category_value == ''):
            
            return 'Not Avaialble'
        
        else: 
            
            return category_value
    
    

    
    
# Function to get the old product price


def get_old_price(i):

        price = soup.findAll('s', {"class": "price-item--regular"})
        price_modified = price[i].text.strip()[3:9]
               
        return price_modified

    
    
    
    
# Function to get the new product price


def get_new_price(i):

        price = soup.findAll('span', {"class": "price-item--sale"})
        price_modified = price[i].text.strip()[3:9]
               
        return price_modified





# Function to get the product rating


def get_rating(i):

        rate = soup.findAll('p', {"class": "rating-text"})
        rate_value = rate[i].text.strip()
        
        rate_modified = round(float(rate_value), 2)
        
        if (rate_modified == 0):
            
            return 'Not Avaialble'
        
        else:
            
            return rate_modified
        
```

```python
# Getting the number of the products on the page


name = soup.findAll("h3", {"class": "product-title"}) 





# Function that loop through all the products and store the products details into a variable


def data():  
    
    d = {"product_name":[], "category":[], "old_price":[], "new_price":[], "rating":[]}
    
    i = 0
    while i < len(name):

            d['product_name'].append(get_name(i)) 
            d['category'].append(get_category(i))
            d['old_price'].append(get_old_price(i))
            d['new_price'].append(get_new_price(i))
            d['rating'].append(get_rating(i))
                                      
            i = i + 1
    
    return d
    
        
    

    
# creating a dataframe to store the data, then storing the dataframe in a csv file


raw_african_offers = pd.DataFrame.from_dict(data())

raw_african_offers.to_csv("raw_african_offers.csv", header=True, index=False)
```
